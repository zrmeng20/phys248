Complete  Activity 1 before the end of the lab by entering the answer into the Brigthspace quiz.

Solve the following differential equation numerically using `scipy.odeint` in the intervall $x \in [0.5,8]$ and $y(0.5) = 2$.
$$
\frac{dy}{dx} = 2 x \sin(x) \, e^\frac{-2x}{3}
$$
and plot $y(x)$. Use an  array for $x$ with $31$ equidistant points.


Multiply the maximum of $y$ and the $x$ value at which $y$ has the maximum, round to two decimals and enter into quiz `max-ode1`.

**Solution:**

```python
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.5, 8, 31)
dydx = lambda y, x: 2*x*np.sin(x)*(np.e**(-2*x/3))
y0 = 2

yy = integrate.odeint(dydx, y0, x)

# find maximum value
ind = np.where(yy==max(yy))[0][0]
res = yy[ind][0]*x[ind]
print('product of yMax and x: {}'.format(res))

plt.figure(4)
plt.title('ODE')
plt.plot(x,yy,'-',label='y(x)')
plt.legend()
plt.show()
```
