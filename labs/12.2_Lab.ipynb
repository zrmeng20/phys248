### Activity 1
Name: `how-many-f.gt.g`, `mean-f`

Define the follow functions as python functions:
$$
f(x) = \sin(0.5+x)^4
$$
and 
$$
g(x) = \sin(\frac{1}{x^3})
$$

1. Make make a plot of both functions for $x \in [0.4,1.5]$. To evaluate the functions use an equidistant numpy array `x` with `n=50` intervals. 
2. Using this `x` array for how many values is $f(x)>g(x)$? Enter the number into quiz `how-many-f.gt.g`.
3. What is the mean of all values $f(x)$ with the array `x` from 1. where $f(x)>0.3$ and $f(x)<g(x)$? Enter the value into quiz `mean-f`.

```python
%pylab ipympl
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 51
x = np.linspace(0.4, 1.5, n)

f = lambda x: np.sin(0.5+x)**4
g = lambda x: np.sin(1/(x**3))

plt.plot(x, f(x))
plt.plot(x, g(x))
plt.show()

res1 = np.where(f(x)>g(x))[0]
print(f'There are {len(res1)} values where f(x) > g(x)')

print(f'The mean of all values f(x)>0.3 and f(x)<g(x): {np.mean(f(x)[np.where((f(x)<g(x))&(f(x)>0.3))[0]]):.2f}')
```

### Activity 2
Name: `lin-reg-slope`, `pearson-r`, `coeff-quadr`

Given are the `x,y` arrays as defined in the code cell below.

1. Make make a plot of `y` vs. `x`.
2. Determine the slope and intercept by linear regression. Enter the slope into quiz  `lin-reg-slope`. Check your values by adding the linear relation with determined slope and intercept to the plot. 
3. Determine the Pearson-r correlation coefficient and enter into quiz `pearson-r`. 
4. Fit a quadratic function $f(x) = ax^2 +bx +c$ to the $x-y$ data. Enter the quadratic coefficient $a$ into quiz `coeff-quadr`. Add the fitted quadratic relation to the plot. 


```python
import matplotlib.pyplot as plt
from scipy import stats
from scipy.optimize import curve_fit

x,y = np.array([[-2.        , -1.1       , -0.2       ,  0.7       ,  1.6       ,
         2.5       ,  3.4       ,  4.3       ,  5.2       ,  6.1       ,
         7.        ],
       [ 2.66642853,  2.26954432,  2.23404689,  5.8540483 ,  6.68994527,
         7.06461448,  6.10310614,  7.2370884 ,  6.92047266,  6.76042916,
         8.68281513]])
```

```python
plt.plot(x, y, 'o')
slope, b, r, p, err = stats.linregress(x, y)
print(f'slope: {slope}')
print(f'intercept: {b}')
plt.plot(x, slope*x+b)

print(f'Pearson-r correlation: {r}')

def fnc(x, a, b, c):
    return a*x**2+b*x+c
y2 = curve_fit(fnc, x, y)
plt.plot(x, fnc(x, y2[0][0], y2[0][1], y2[0][2]))
print(f'coefficient a: {y2[0][0]}')
plt.show()
```
